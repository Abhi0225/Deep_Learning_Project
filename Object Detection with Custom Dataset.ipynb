```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame=cap.read()
    
    result=model(frame)
    
    cv2.imshow('yolo', np.squeeze(result.render()))
    
    if cv2.waitKey(10) & 0XFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
