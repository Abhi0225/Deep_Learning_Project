```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
model
```

```python
img = 'https://www.enconassociates.com/i/commute_2225803c.jpg'
```

```python
result = model(img)
result.print()
```

```python
%matplotlib inline
plt.imshow(np.squeeze(result.render()))
plt.show()
```

```python
result.show()
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame=cap.read()
    
    result=model(frame)
    
    cv2.imshow('yolo', np.squeeze(result.render()))
    
    if cv2.waitKey(10) & 0XFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
        
```
